```python
import matplotlib.pyplot as plt
import numpy as np
```

# plot psf versus measured speckle sizes

```python
measured_size_NA=[710,273,221,170]
psf_NA=[396,223,137,127]
n=['0.45 N.A.','0.8 N.A.', '1.3 N.A.', '1.4 N.A.']
fig, ax = plt.subplots()
ax.scatter(psf_NA, measured_size_NA,c='k')

for i, txt in enumerate(n):
    ax.annotate(txt, (psf_NA[i]-10, measured_size_NA[i]-30),xycoords='data')

plt.tick_params(bottom=True, top=True, left=True, right=True,direction='in')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylim(0,800)
plt.xlim(0,500)
plt.xlabel('Point Spread Function (nm)',size=18)
plt.ylabel('Speckle Size (nm)',size=18)
plt.legend(fontsize=18)
```

```python
measured_size_NA=[648,710,755,779]
psf_NA=[333,396,423,450]
n=['0.26 A.U.','0.52 A.U.', '0.78 A.U.', '1 A.U.']
fig, ax = plt.subplots()
ax.scatter(psf_NA, measured_size_NA,c='k')

for i, txt in enumerate(n):
    ax.annotate(txt, (psf_NA[i]-6, measured_size_NA[i]+5),xycoords='data')

plt.tick_params(bottom=True, top=True, left=True, right=True,direction='in')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.xlabel('Point Spread Function (nm)',size=18)
plt.ylabel('Speckle Size (nm)',size=18)
plt.legend(fontsize=18)
```
