```python
with open('/Volumes/wwyyrx/Wenyun/Silk_speckle/analysis/fresh_linescan_g2t.pkl', 'rb') as file:
    fresh_linescan_g2t = pickle.load(file)

with open('/Volumes/wwyyrx/Wenyun/Silk_speckle/analysis/fresh_trc.pkl', 'rb') as file:
    trc_fresh = pickle.load(file)


with open('/Volumes/wwyyrx/Wenyun/Silk_speckle/analysis/10min_trc.pkl', 'rb') as file:
    trc_10min = pickle.load(file)

with open('/Volumes/wwyyrx/Wenyun/Silk_speckle/analysis/20min_trc.pkl', 'rb') as file:
    trc_20min = pickle.load(file)

with open('/Volumes/wwyyrx/Wenyun/Silk_speckle/analysis/30min_trc_fastscan.pkl', 'rb') as file:
    trc_30min = pickle.load(file)
```

```python
fsfresh_line=0.00023
fsfresh=0.03256
fs10min=0.02949
fs20min=0.03256
fs30min=0.03256
```

```python
trc_10min.shape
```

```python
laglist_fresh=range(0,int(15.0/fsfresh),1)
laglist_10min=range(0,int(15.0/fs10min),1)
laglist_20min=range(0,int(15.0/fs20min),1)
laglist_30min=range(0,int(15.0/fs30min),1)
lagtime_fresh= [element * fsfresh for element in laglist_fresh]
lagtime_10min= [element * fs10min for element in laglist_10min]
lagtime_20min= [element * fs20min for element in laglist_20min]
lagtime_30min= [element * fs30min for element in laglist_30min]

fig, ax = plt.subplots(figsize=(6, 5))

ax.plot(lagtime_fresh,trc_fresh,'k.-',label='0 min')


ax.plot(lagtime_10min,trc_10min,'g.-',label='10 min')

ax.plot(lagtime_20min,trc_20min,'b.-',label='20 min')

ax.plot(lagtime_30min,trc_30min,'r.-',label='30 min')

ax.legend()
ax.tick_params(direction='in', length=6, width=2, colors='k',
               which='major', labelsize=14, top=True, right=True, bottom=True, left=True)
ax.set_xlabel("Lagtime",fontsize=18)
ax.set_ylabel("Correlation",fontsize=18)
ax.tick_params(direction='in')
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_ylim(0,1)
```
